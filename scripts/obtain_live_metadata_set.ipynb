# From the Merged Metadata produce the final set in a csv that is to be loaded into DB


```python
import pandas as pd
```

Load in the csv. This is the product of the metadata cleaning scripts. It has x, y and st variants of core columns. 

```python
csv_file = '../data/Cleaned_Metadata_Merged_w_Manual_x_y_st_metadata_merged_with_manual.csv'
csv_file = "../standardized_columns_final_2023-12-18.csv"
```

```python
df = pd.read_csv(csv_file)
df.head()
```

Write these values into a df with all columns (including those from SRA)

```python
full_df = pd.read_csv(csv_file)
full_df.head()
```

```python
update_columns = [col for col in full_df.columns if col.endswith("_st")]

for col in update_columns:
    base_col = col[:-3]
    full_df[base_col] = df[col]
full_df.head()
```

write the df to file

```python
full_df.drop(columns=["name", 'not_unique', 'Run.1', 'BioProject.1'], inplace=True)
full_df.drop(columns=update_columns, inplace=True)
```

```python
full_df.drop(columns=["Experimental Factor: wild type genotype"], inplace=True)
```

```python
from datetime import datetime
today_date = datetime.today().strftime('%Y-%m-%d')
file_name = f'../data/Cleaned_Metadata_For_Upload_{today_date}.csv'

full_df.to_csv(file_name, index=False)
```
